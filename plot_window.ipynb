```python
from tkinter import *
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import pandas as pd
import numpy as np
import matplotlib.animation as animation
```

```python
data=pd.DataFrame(pd.read_csv('CanSat_DHT11.csv')).dropna()
data
```

```python
temperature=data['Temp C']
temperature_plot=plt.figure(figsize=(20,10))
plt.plot(temperature,linestyle='-',marker='o',ms=10,mec='r',mfc='red')
plt.title('Temperature',fontsize=20)
plt.xlabel('Time (s)',fontsize=20)
plt.ylabel('Temperature `C',fontsize=20)
plt.show()
```

```python
humidity=data['Humidity %']
humidity_plot=plt.figure(figsize=(20,10))
plt.plot(humidity,linestyle='-',marker='o',ms=10,mec='r',mfc='red')
plt.title('Humidity',fontsize=20)
plt.xlabel('Time (s)',fontsize=20)
plt.ylabel('Humidity %',fontsize=20)
plt.show()
```

```python
heat_index=data['HeatIndex C']
heat_index_plot=plt.figure(figsize=(20,10))
plt.plot(heat_index,linestyle='-',marker='o',ms=10,mec='r',mfc='red')
plt.title('Heat Index',fontsize=20)
plt.xlabel('Time (s)',fontsize=20)
plt.ylabel('Heat Index C',fontsize=20)
plt.show()
```

```python
# plotting all 3 graphs in same figure
# figure,ax=plt.subplots(nrows=1,ncols=1,figsize=(20,10))
figure=plt.figure(figsize=(20,10))
plt.plot(temperature,linewidth=5)
plt.plot(humidity,linewidth=5)
plt.plot(heat_index,linewidth=5)
plt.legend(['Temperature','Humidity','HeatIndex'],loc='lower right',fontsize=20,facecolor='yellow',shadow=True,prop={'size':30})
```
